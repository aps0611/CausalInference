Machine Learning is currently very good at answering the type of question of the prediction kind

ML is notoriously bad at this inverse causality type of problem. They require us to answer “what if” questions, which economists call counterfactuals. What would happen if I used another price instead of this price I’m currently asking for my merchandise?

❎ ASSOSIATION IS NOT CAUSATION

## WHEN ASSOSIATION IS CAUSATION

```python
import pandas as pd
import numpy as np
from scipy.special import expit
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
np.random.seed(123)
n = 100 # number of data points we need for dataset
tuition = np.random.normal(1000, 300, n).round() # Generates the n data points with mean 1000 and std() of 300
tablet = np.random.binomial(1, expit((tuition - tuition.mean()) / tuition.std())).astype(bool) # generates the binomial value for each tuition datapoint based on expit function
enem_score = np.random.normal(200 - 50 * tablet + 0.7 * tuition, 200) # enem score with standard deviation set to 200
enem_score = (enem_score - enem_score.min()) / enem_score.max()
enem_score *= 1000 # change the scale from [0,1] to [0,1000]

data = pd.DataFrame(dict(enem_score=enem_score, Tuition=tuition, Tablet=tablet))
```

```python
data
```

```python
plt.figure(figsize=(6,8))
sns.boxplot(y="enem_score", x="Tablet", data=data).set_title('ENEM score by Tablet in Class')
plt.show()
```
